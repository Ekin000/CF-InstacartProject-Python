## If-statement with user-defined function

```python
# Define a function
```

```python
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# Apply the function
```

```python
df_ords_prods_merged['price_range'] = df_ords_prods_merged.apply(price_label, axis=1)
```

```python
def busiest_days(row):

  if row['order_dow'] <= 1:
    return 'Busiest days'
  elif row['order_dow'] == 2:
    return 'Regularly busy'
  elif (row['order_dow'] >= 3) and (row['order_dow'] < 5):
    return 'Least busy'
  elif row['order_dow'] >=5:
    return 'Regularly busy'
  else: return 'Not enough data'
```

```python
df_ords_prods_merged['busiest_days'] = df_ords_prods_merged.apply(busiest_days, axis=1)
```

```python
def busiest_period(row):

  if (row['order_time'] >= 9) and (row['order_time'] <=16):
    return 'Most orders'
  elif (row['order_time'] >= 7) and (row['order_time'] <=8):
    return 'Average orders'
  elif (row['order_time'] >= 17) and (row['order_time'] <= 22):
    return 'Average orders'
  elif (row['order_time'] <= 6):
    return 'Fewest orders'
  elif row['order_time'] == 23:
    return 'Fewest orders'
  else: return 'Not enough data'
```

```python
df_ords_prods_merged['busiest_period_of_day'] = df_ords_prods_merged.apply(busiest_period, axis=1)
```
