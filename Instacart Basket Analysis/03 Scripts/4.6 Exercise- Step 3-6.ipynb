# Step-6

```python
# to check if the matching is correct. 
df_combined.merge(df_prods, on='product_id', how='outer', indicator='exists')
```

```python
(df_combined.merge(df_prods, on='product_id', how='outer', indicator='exists')['exists']).value_counts()
```

```python
(df_combined.merge(df_prods, on='product_id', how='outer', indicator='exists')['exists']).value_counts(dropna=False)
```

```python
# with how_outer argument, it is seen that the first matching wasn`t a full match. 
```

```python
# to create a full match
df_full_merged=df_combined.merge(df_prods, on='product_id', how='outer', indicator='exists')
```

```python
df_full_merged['_merge'].value_counts(dropna=False)
```
