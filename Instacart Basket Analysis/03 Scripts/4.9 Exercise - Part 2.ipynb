## Sampling Data

```python
# to create a list holding True/False values to the test np.random.rand() <= 0.7
np.random.seed(4)
dev=np.random.rand(len(df)) <= 0.7
```

```python
dev
```

```python
# to store 70% of the sample in the dataframe big
big=df[dev]
```

```python
# to store 30% of the sample in the dataframe small
small=df[~dev]
```

```python
len(df)
```

```python
len(big) + len(small)
```

```python
# we don`t need the all columns for the line chart
df_2=small[['order_dow', 'prices']]
```

```python
df_2.head()
```

```python
# to create the line chart of small dataframe
line=sns.lineplot(data=df_2, x='order_dow', y='prices')
```
